```python
lemmatizer=WordNetLemmatizer()
```

```python
def get_simple_pos(tag):

  if tag.startswith('J'):
    return wordnet.ADJ
  elif tag.startswith('V'):
    return wordnet.VERB
  elif tag.startswith('R'):
    return wordnet.ADV
  else:
    return wordnet.NOUN

```

```python
# w='better'
# pos_tag([w])

# Remember that pos_tag expects a list as its argument
```

```python
# Create a list of stopwords
stop_words=stopwords.words("english")

# Create a list of all punctuation marks
punctuations = list(string.punctuation)

# Use them collectively
stop_words=stop_words+punctuations
```

```python
def clean_review(textFile):
  output_words=[]

  for word in textFile:
    if word.lower() not in stop_words:
      pos=pos_tag([word]) #word passed as a list
      clean_word=lemmatizer.lemmatize(word,pos=get_simple_pos(pos[0][1]))#Part of speech is (word,pos)
      output_words.append(clean_word.lower())
  return output_words
```

```python
# documents= [(clean_review(document), category) for document,category in documents]

for i, (document, category) in enumerate(documents):
    documents[i] = (clean_review(document), category)
```

```python
training_documents=documents[0:1500]
testing_documents=documents[1500:]
```

```python
all_words=[]

# word_list=((list of words),category)

for word_list in training_documents:
    all_words.extend(word_list[0])


# all_words
```

```python
# Find frequency distribution of these words
freq=nltk.FreqDist(all_words)
```

```python
common=freq.most_common(3000)
# common
```

```python
features=[w[0] for w in common]
# features
```

```python
documents[0]
```

```python
# Function to create a feature dictionary 

def get_feature_dictionary(words):
    current_features={}
    word_set=set(words)

    for w in features:
        current_features[w]=w in word_set

    return current_features
    
```

```python
output=get_feature_dictionary(training_documents[0][0])
output
```

```python
training_data=[(get_feature_dictionary(doc),category) for doc,category in training_documents]
testing_data=[(get_feature_dictionary(doc),category) for doc,category in testing_documents]
```

```python
training_data[0]
```

```python
classifier=NaiveBayesClassifier.train(training_data)
```

```python
nltk.classify.accuracy(classifier,testing_data)
```
